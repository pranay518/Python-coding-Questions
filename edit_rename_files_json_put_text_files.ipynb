```python
# Create a random file name 
def createrandname(path,ext):
	check=False
	while check == False:
		rand="".join(random.sample(string.ascii_lowercase+string.digits,20))+ext
		check=True if not os.path.exists(path+rand) else False
	return rand
```

```python
try:
    for this in os.listdir(path):
            print("Renaming:", this)
            print("this is {}".format(this))
            try:
                if os.path.isfile(path + "/" + this):
                    rand=createrandname(path,os.path.splitext(this)[1])
                    #Do the rename
                    os.rename(path + "/" + this, path +"/" + rand)
                    print("\t->", rand)
                    with open(via_file_path, 'r') as file:
                        content = file.read()
                        content=content.replace(this, rand)
                    with open(via_file_path, 'w') as f:   
                        f.write(content)
                        print("JSON Data replaced")
                    with open (txt_file_path + '/'+ "renameFile.txt", 'a') as text_file:
                        text_file.write(this +"\t->"  + rand + "\n" )
                else:
                    #Not a file
                    print("\tSkipped:", "'"+this+"'", "Target is not a file")
            except:
                print("Error might be due to rename functionality or via file content replacement or write functionality in text file")
    print("Finished!")
except:
    print("path error")
```
